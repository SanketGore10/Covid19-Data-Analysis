# Q. 3) In which Region, maximum number of Confirmed cases were recorded ?


```python
data
```

```python
data["Confirmed"].groupby(data.Region).sum().sort_values(ascending=False)
```

# Q. 4) In which Region, minimum number of Deaths cases were recorded ?


```python
data["Deaths"].groupby(data.Region).min().sort_values()
```
